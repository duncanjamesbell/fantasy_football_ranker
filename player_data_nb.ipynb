```python
#collect position score data for each league, save to file for each year
for league_url in league_urls:
    #go to player page, filter to a position
    player_url = league_url + '/players'
    from selenium.webdriver.support.ui import Select
    driver.get(player_url)
    time.sleep(1)

    status_selector = Select(driver.find_element(By.XPATH,"//select[@id='statusselect']"))
    status_selector.select_by_visible_text('All Players')
    time.sleep(1)

    #assemble list of available positions we could select
    position_options = []
    for o in driver.find_element(By.XPATH,"//select[@id='posselect']").find_elements(By.TAG_NAME,'option'):
        position_options.append(o.text)

    full_position_dfs = []
    for position in ['QB','WR','RB','TE','K','DEF']:
        if position in position_options:
            print(f'Collecting {position} players...')
            pos_selector = Select(driver.find_element(By.XPATH,"//select[@id='posselect']"))
            pos_selector.select_by_visible_text(position)
            time.sleep(1)
            driver.find_element(By.XPATH,"//input[@data-slk='players_filter_click']").click()
            time.sleep(3)
            #SCRAPING PLAYER DATA
            more_available = True

            position_dfs = []
            while more_available:
                player_elems = driver.find_elements(By.XPATH,"//a[@class='Nowrap name F-link']")
                rank_elements =driver.find_elements(By.CSS_SELECTOR,"td[class^='Alt Ta-end Selected']")
                player_names = []
                player_urls = []
                for p in player_elems:
                    player_names.append(p.text)
                    player_urls.append(p.get_attribute('href'))

                player_page_count = len(player_elems)
                ranks = []
                for s in rank_elements:
                    try:
                        ranks.append(int(s.text))
                    except:
                        pass

                #correct mismatch of ranks / players
                if len(ranks) < player_page_count:
                    count_match = False
                    add=0
                    while count_match == False:
                        ranks.append(99900 + add)
                        if len(ranks) < player_page_count:
                            add=add+1
                        else:
                            count_match = True

                player_df = pd.DataFrame()
                player_df['player_name'] = player_names
                player_df['position'] = position
                player_df['player_url'] = player_urls
                player_df['position_rank'] = ranks

                position_dfs.append(player_df)

                try:
                    next_page = driver.find_element(By.XPATH,"//div[@class='pagingnav navlist']").find_elements(By.TAG_NAME,'a')[-1]
                    if next_page.text == 'Next 25':
                        driver.get(next_page.get_attribute('href'))
                    else:
                        more_available = False
                except:
                    more_available = False

            full_position_df = pd.concat(position_dfs)
            full_position_dfs.append(full_position_df)
    all_positions = pd.concat(full_position_dfs)
    year = league_url.split('https://football.fantasysports.yahoo.com/')[1].split('/')[0]
    position_filename = f'{year}_position_scores.csv'
    all_positions.to_csv(position_filename)
```

```python
#building a player master list - USING API, which has limitations...seems we can only use taken_players
player_dfs = []
player_dict = {}
for league in leagues:
    l = football.to_league(league)
    players = l.taken_players()
    for p in players:
        if p['player_id'] not in player_dict:
            player_dict[p['player_id']] = (p['name'],p['eligible_positions'])
            player_df = pd.DataFrame(p)
            player_df['league'] = league
            player_dfs.append(player_df)
taken_players = pd.concat(player_dfs)
taken_players.reset_index(drop=True,inplace=True)

#handling for position duplicates
flex_positions = taken_players[taken_players.eligible_positions.isin(['Q/W/R/T','W/R/T','W/T'])]
normal_positions = taken_players[~taken_players.eligible_positions.isin(['Q/W/R/T','W/R/T','W/T'])]
normal_positions = normal_positions.drop_duplicates(subset=['name','eligible_positions'])
missing_flex_df = flex_positions[~flex_positions.name.isin(normal_positions.name)].copy()
missing_flex_df.sort_values(by='eligible_positions',ascending=True,inplace=True)
missing_flex_df.drop_duplicates(subset=['player_id'],inplace=True)
player_master = pd.concat([normal_positions,missing_flex_df])
#there remain a small number of duplicates, players like Ty Montgomery with multiple position types
#normal_positions[normal_positions.name=='Ty Montgomery']
player_master
```

```python
#alright, we need to reconcile the master player list that has been assembled in name_dict
#with the full list of player names that were started in regular season and playoffs
#and, actualy, also transactions.  I should do that first.

#ok let's compile a list of all players from regular season starts, playoff starts, and faab pickups

regular_season = pd.read_csv('all_regular_season_thru_2022.csv')
rs_players = regular_season[regular_season.player_id.notnull()].drop_duplicates(subset=['player_id'])

playoffs = pd.read_csv('all_playoffs_thru_2022.csv')
playoff_players = playoffs[playoffs.player_id.notnull()].drop_duplicates(subset=['player_id'])

faab = pd.read_csv('faab_thru_2022.csv')
faab_players = faab[faab.player_id.notnull()].drop_duplicates(subset=['player_id'])

player_dict_master = pd.read_csv('names_dict.csv')[['player_name','position','player_id']].drop_duplicates()

player_dict = {}
for index, row in player_dict_master.iterrows():
    player_dict[row['player_id']] = (row['player_name'],row['position'])

print(f'Original len: {len(player_dict)}')

#add faab first because we prefer full names
for index, row in faab_players.iterrows():
    if row['player_id'] not in player_dict:
        player_dict[row['player_id']] = (row['player_name'],row['position'])

print(f'After faab: {len(player_dict)}')

#add regular season players
for index, row in rs_players.iterrows():
    if row['player_id'] not in player_dict:
        player_dict[row['player_id']] = (row['player'],row['position'])

print(f'After regular season: {len(player_dict)}')

#add playoff players
for index, row in playoff_players.iterrows():
    if row['player_id'] not in player_dict:
        player_dict[row['player_id']] = (row['player'],row['position'])

print(f'After playoffs: {len(player_dict)}')

player_dict
```
